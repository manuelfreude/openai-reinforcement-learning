```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
results_DDPG_keras = pd.read_csv('data_DDPG_keras.csv')
results_DDPG_keras['Episode']=results_DDPG_keras.reset_index().index
results_DDPG_keras.head()
```

```python
results_DDPG_keras.drop(['reward_DDPG'], axis=1)
```

```python
results_DDPG_keras.columns = ['Reward DDPG keras', 'Episode_del', 'Episode']
```

```python
results_DDPG_keras.drop(['Episode_del'], axis=1, inplace=True)
```

```python
results_DDPG_keras.head()
```

```python
results_DDPG_keras.tail()
```

```python
results_DDPG_keras.info()
```

```python
results_DDPG_keras['Reward DDPG keras'] = results_DDPG_keras['Reward DDPG keras'].str[1:]
results_DDPG_keras.head()
```

```python
results_DDPG_keras['Reward DDPG keras'] = results_DDPG_keras['Reward DDPG keras'].str[:-1]
results_DDPG_keras.head()
```

```python
results_DDPG_keras['Reward DDPG keras'] = results_DDPG_keras['Reward DDPG keras'].astype(float, inplace=True)
```

```python
results_DDPG_keras.head()
```

```python
results_DDPG_keras.info()
```

```python
results_DDPG_keras_200 = results_DDPG_keras[:200]
```

```python
plt.plot(results_DDPG_keras_200['Episode'], results_DDPG_keras_200['Reward DDPG keras'])
plt.ylabel('Rewards per Episode')
plt.xlabel('Episode')
plt.legend(['Reward DDPG keras']) #plt.legend(['Reward PSA', 'Reward DDPG'])
_ = plt.ylim();
```
